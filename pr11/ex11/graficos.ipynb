```python
arquivos = ['100000-contrario.txt', '100000-misturado.txt',
            '100000-ordenado.txt']
condicao = ['contrario', 'misturado', 'ordenado']
algoritmos = [1, 12, 2, 21, 3, 32, 4, 42]
nomeAlgoritmo = ['shell', 'shellR', 'quick', 'quickR', 'heap', 'heapR', 'merge', 'mergeR']

fig, axs = plt.subplots(8, 3, figsize=(18*2, 12*3))  # Set the figure size

for j, algoritmos in enumerate(algoritmos):
    for i, arquivo in enumerate(arquivos):
        nome_saida = f"res_{nomeAlgoritmo[j]}_{condicao[i]}.csv"
        
        df = pd.read_csv(nome_saida, names = ['n', 'tempo'])

        df['n'] = pd.to_numeric(df['n'])
        df['tempo'] = pd.to_numeric(df['tempo'])

        axs[j, i].plot(np.array(df['n']), np.array(df['tempo']))  # Plotting x and y data
        axs[j, i].set_ylabel("Tempo (s)")  # Set y-axis label
        axs[j, i].set_xlabel("n")  # Set x-axis label
        axs[j, i].set_title(f"{nomeAlgoritmo[j]} {condicao[i]}")  # Set subplot title

plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()  # Show the plot
    
        
```
